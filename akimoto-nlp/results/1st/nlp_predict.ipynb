```python
BASE_DIR = '/content/drive/MyDrive/Colab Notebooks/akimoto/nlp'
TYPE = 'gru'
```

```python
from keras.models import model_from_json

n_rnn = 10

json_string = open(f"{BASE_DIR}/{TYPE}_model.json").read()
model = model_from_json(json_string)

model.summary()

model.compile(loss='categorical_crossentropy', optimizer="adam")

model.load_weights(f"{BASE_DIR}/{TYPE}_weights.hdf5")
```

```python
import numpy as np

# これ、知ってる必要あるんだっけ？取ってこれる？
# char_indexes もいりそう。。。
chars = []
with open(f"{BASE_DIR}/chars.json", mode='r') as f:
    chars = list(map(lambda s: s.strip(), f.readlines()))

    chars.pop(1)
    chars[0] = '\n'

char_indices = {}
for i, char in enumerate(chars):
    char_indices[char] = i
indices_char = {}
for i, char in enumerate(chars):
    indices_char[i] = char
n_chars = len(chars)
print(n_chars)
print(chars)
print(char_indices)
print(char_indices['こ'])

def generate_text(input):

    beta = 5  # 確率分布を調整する定数
    prev_text = input[:n_rnn]  # 入力に使う文字
    created_text = prev_text  # 生成されるテキスト
    
    print("シード: ", created_text)

    for i in range(400):
        # 入力をone-hot表現に
        x_pred = np.zeros((1, n_rnn, n_chars))
        for j, char in enumerate(prev_text):
            x_pred[0, j, char_indices[char]] = 1
        
        # 予測を行い、次の文字を得る
        y = model.predict(x_pred)
        p_power = y[0] ** beta  # 確率分布の調整
        next_index = np.random.choice(len(p_power), p=p_power/np.sum(p_power))        
        next_char = indices_char[next_index]

        created_text += next_char
        prev_text = prev_text[1:] + next_char

    print(created_text)
    print()
```

```python
generate_text('こんにちは、今日はいい天気ですね')
```
